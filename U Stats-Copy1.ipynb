```python
#File: U Stats - Sheet1.csv
#LINK: https://www.nber.org/system/files/working_papers/w23618/w23618.pdf

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from matplotlib import font_manager
```

```python
#Set style for rest of notebook
matplotlib.style.use("default") #for bar charts
#matplotlib.style.use("classic") #for line graphs
matplotlib.rcParams["font.size"] = 13
```

```python
#Function for percentages in pie chart from: https://stackoverflow.com/questions/21572870/matplotlib-percent-label-position-in-pie-chart
def func(pct):
  return "{:1.1f}%".format(pct)
```

```python
#Creating the pie chart, had to create individual data framnes for each pie chart because the data came from a table that was in a non-exportable pdf
#I had to manually transfer the data to the data frames but it works properly

df = pd.DataFrame({'-': [3.8, 4.3, 5.5, 4.1, 8.4, 7.1, 17, 10.7, 14.6, 21.1, 20.6, 20.9]},
index=["Ivy Plus ", "Other Elite Colleges ", "Highly Selective Public ", "Highly Selective Private ", "Selective Public", "Selective Private", "Nonsel 4-year Public", "Nonsel 4-year Private ", '2-year Non-profit', '4-year For-profit', '2-year For-profit ', 'Less than 2-year Colleges'])

df.plot.pie(y='-', figsize=(10,10), legend = False, autopct=lambda pct: func(pct))

plt.savefig("1PiBottom20%.png", dpi = 300)
#Bottom 20% pie chart
```
