```python
#File:salaries-by-college-type.csv
#Link: https://osf.io/db4g8

#Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import matplotlib
from matplotlib import font_manager
```

```python
#File reader
file = "salaries-by-college-type.csv"
df = pd.read_csv(file)
print(df.head())
print(list(df))
```

```python
#Set style for rest of notebook
#matplotlib.style.use("fivethirtyeight") #for bar charts
matplotlib.style.use("classic") #for line graphs
matplotlib.rcParams["font.size"] = 10
```

```python
#Frequency Chart
col = 'School Type'
col_name = 'School Type'
freq = df[col].value_counts().rename_axis(col_name).reset_index(name='# of Universities')
freq = freq.sort_values(by = col_name)

ax = freq.plot.barh(x=col_name, y='# of Universities', rot=0)
```

```python
#Create groupbys and means
type_mean = df.groupby("School Type")["Starting Median Salary"].mean()
type1_mean = df.groupby("School Type")["Mid-Career Median Salary"].mean()
type2_mean = df.groupby("School Type")["Mid-Career 10th Percentile Salary"].mean()
type3_mean = df.groupby("School Type")["Mid-Career 25th Percentile Salary"].mean()
type4_mean = df.groupby("School Type")["Mid-Career 75th Percentile Salary"].mean()
type5_mean = df.groupby("School Type")["Mid-Career 90th Percentile Salary"].mean()
```

```python
#Graph with abr chart and Line graphs
fig2 = plt.figure(figsize=(9,7))

type_mean.plot(kind = "line", title = "Starting Salary by College Type", ylabel = "Starting Median Salary", xlabel = "School Type", color = "Black", linewidth= 2)
type1_mean.plot(kind = "bar", title = "Mid-Career Median Salary by College Type", ylabel = "Mid-Career Median Salary", xlabel = "College Type", color = "SkyBlue")
type2_mean.plot(kind = "line", title = "Mid-Career 10th percentile Salary by College Type", ylabel = "Mid-Career 10th Percentile Salary", xlabel = "College Type", color = "Red", linewidth= 2)
type3_mean.plot(kind = "line", title = "Mid-Career 25th percentile Salary by College Type", ylabel = "Mid-Career 25th Percentile Salary", xlabel = "College Type", color = "Yellow", linewidth= 2)
type4_mean.plot(kind = "line", title = "Mid-Career 75th percentile Salary by College Type", ylabel = "Mid-Career 75th Percentile Salary", xlabel = "College Type", color = "Green", linewidth= 2)
type5_mean.plot(kind = "line", title = "Mid-Career 90th percentile Salary by College Type", ylabel = "Mid-Career 90th Percentile Salary", xlabel = "College Type", color = "Orange", linewidth= 2)

plt.legend() 
```
